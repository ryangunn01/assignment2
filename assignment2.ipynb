```python
#Q4

#Part 1
import pandas as pd
import numpy as np
df = pd.read_csv('./airbnb_hw.csv')
df['Price'] = df['Price'].str.replace('$','')
df['Price'] = df['Price'].str.replace(',','')
#Coerce to numeric
df['Price'] = pd.to_numeric(df['Price'],errors='coerce')
#Count missing values
df['Price_nan'] = df['Price'].isnull()

#Part 2
display(df.head())
print(df.shape)
#14 columns with 30,478 rows

#Part 3
pd.crosstab(df['Room Type'],df['Property Type'])
#Apartments are the most common listing. I
#Interesting that tent is listed as private room - is it just considered differently from traditional units like house/apartment?
#Private rooms are most common in apartments. More likely for listing to have a private room versus entire unit at a bed and breakfast, house.

#Part 4
import seaborn as sns
#Need to change scale
df['Price_arcsinh'] = np.arcsinh(df['Price'])
sns.histplot(df['Price_arcsinh'])
sns.kdeplot(df['Price_arcsinh'])
sns.boxplot(data=df,x='Price')
#There are a lot of outliers beyond the max. There is a normal distribution.
df['Price_log'] = np.log(df['Price'])
sns.histplot(df['Price_log'])
sns.kdeplot(df['Price_log'])
sns.boxplot(data=df,x='Price_log')
#Logarthimic transformation gives similar results to arcsinh transformation.
sns.scatterplot(data=df,x='Beds',y='Price_log')
#Not linear. Number of properties lessens with increase in beds.

#Part 5
group_mean = df.groupby('Beds')['Price_log'].mean().reset_index()
group_std = df.groupby('Beds')['Price_log'].std().reset_index()
sns.scatterplot(data=group_mean,x='Beds',y='Price_log')
sns.scatterplot(data=group_std,x='Beds',y='Price_log')
#There is more of a relationship. With mean: almost s-curved. Standard deviation graph is similar.

#Part 6
sns.scatterplot(data=df,x='Beds',y='Price_log',hue='Room Type', palette='viridis')
sns.scatterplot(data=df,x='Beds',y='Price_log',hue='Property Type', palette='viridis')
grouped = df.groupby(['Room Type','Property Type'])['Price'].describe()
display(grouped)      

#The highest mean price is in 'other', but there are few properties and a high standard deviation. Entire home/apt has the highest mean price.          
#The median is probably better suited for analysis of central tendency, as there are outliers and the mean is sensitive to outliers.          
                 
#Part 7
sns.jointplot(data=df,x='Beds',y='Price_log',kind='hex')
# The data are most densely distributed to the left of the jointplot, where the hexagons are darkest. This is not clear when looking at the previous plots. 
# Because of this, it is important to analyze and visualize your data from many angles.
```
